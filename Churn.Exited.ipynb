```python
### removing the unnecessary columns

df = df[['CreditScore', 'Geography',
       'Gender', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard',
       'IsActiveMember', 'EstimatedSalary', 'Exited']]
```

```python
df.head()
```

```python
### importing the labelencoder from sklearn to convert the categorical to numerical

from sklearn.preprocessing import LabelEncoder
```

```python
### fitting the columns  to convert the categorical to numerical

le = LabelEncoder()
df['Geography'] = le.fit_transform(df['Geography'])
```

```python
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
```

```python
df
```

```python
### giving the x and y 

x = df[['CreditScore', 'Geography',
       'Gender', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard',
       'IsActiveMember', 'EstimatedSalary',]].values
y = df[['Exited']].values
```

```python
### importing the PCA from sklearn

from sklearn.decomposition import PCA
```

```python
### fitting the x values

pca = PCA()
pca.fit_transform(x)
vals = pca.explained_variance_ratio_
```

```python
vals
```

```python
import matplotlib.pyplot as plt
plt.pie(vals)
```

```python
vals * 100
```

```python
### importing the train_test_split from sklearn

from sklearn.model_selection import train_test_split
```

```python
### giving the x_train, x_test, y_train, y_test 

x_train, x_test, y_train, y_test = train_test_split(x,y)
```

```python
### importing the KNN from sklearn 

from sklearn.neighbors import KNeighborsClassifier , KNeighborsRegressor
```

```python
knn = KNeighborsClassifier(n_neighbors = 5)
```

```python
knn
```

```python
### fitting the x_train, y_train

knn.fit(x_train, y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
### importing the accuracy score from sklearn matrices

from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
